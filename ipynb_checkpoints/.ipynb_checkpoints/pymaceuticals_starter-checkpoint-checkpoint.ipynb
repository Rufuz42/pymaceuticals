## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 
# Creates an empty dataframe to be populated with the summary data
summary_data_df = pd.DataFrame()

# Groups the data by drug regimen so we can apply the math by each drug
grouped_cleaned_df = cleaned_mouse_study_df.groupby("Drug Regimen")

# mean, median, variance, standard deviation, and SEM of the tumor volume. 
summary_data_df["Mean Tumor Volume"] = grouped_cleaned_df["Tumor Volume (mm3)"].mean()
summary_data_df["Median Tumor Volume"]  = grouped_cleaned_df["Tumor Volume (mm3)"].median()
summary_data_df["Tumor Volume Variance"]  = grouped_cleaned_df["Tumor Volume (mm3)"].var()
summary_data_df["Tumor Volume Std. Dev."]  = grouped_cleaned_df["Tumor Volume (mm3)"].std()
summary_data_df["Tumor Volume Std. Err."]  = grouped_cleaned_df["Tumor Volume (mm3)"].sem()

# Assemble the resulting series into a single summary dataframe.
# Already assumed it when we calculated it above
summary_data_df
```

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# Using the aggregation method, produce the same summary statistics in a single line
grouped_cleaned_df2 = cleaned_mouse_study_df.groupby("Drug Regimen").agg({"Tumor Volume (mm3)":["mean","median","var","std","sem"]})
grouped_cleaned_df2
```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin

# Start by getting the last (greatest) timepoint for each mouse
# Create a groupby that stores the max of all columns by Mouse ID. 
max_values_by_mouseID = cleaned_mouse_study_df.groupby("Mouse ID").max()
# Extract just the timepoint column from the above dataframe
max_timepoint = max_values_by_mouseID[["Timepoint"]]
# max_timepoint


# Merge this group df with the original dataframe to get the tumor volume at the last timepoint
# Using merge with max timepoint and cleaned DF and inner to only extract matching columns. 
final_tumor_size_df = pd.merge(max_timepoint, cleaned_mouse_study_df, how="inner", on=["Mouse ID", "Timepoint"])

final_tumor_size_df
```

```python
# Put treatments into a list for for loop (and later for plot labels)
# Puts the treatments into a list and creates an empty DF. 
treatment_list = ["Capomulin", "Ramicane", "Infubinol", "Ceftamin"]
treatments = {}


# Create empty list to fill with tumor vol data (for plotting)
tumor_vol_list = []

#creating a new dataframe holding only the four drugs above
# Creates a DF that takes the tumor volume for each of the 4 drugs above and puts them into a column. Since the drugs
# have the same number of uses, the table has clean values. 
for treatment in treatment_list:
    treatments[f'{treatment}'] = final_tumor_size_df.loc[final_tumor_size_df['Drug Regimen']== treatment,'Tumor Volume (mm3)'].tolist()
    
select_treatments_df = pd.DataFrame(treatments)



# Calculate the IQR and quantitatively determine if there are any potential outliers.
upper_q_df = {}
lower_q_df = {}

for treatment in treatment_list:
    # Locate the rows which contain mice on each drug and get the tumor volumes
    quartiles = final_tumor_size_df.loc[final_tumor_size_df['Drug Regimen'] == treatment, :]['Tumor Volume (mm3)'].quantile([.25,.5,.75])
    # add subset 
    first_q = quartiles[0.25]
    third_q = quartiles[0.75]
    iqr = third_q-first_q
    # Determine outliers using upper and lower bounds
    lower_bound = first_q - (1.5*iqr)
    upper_bound = third_q + (1.5*iqr)
    # Stores the upper and lower bound for each treatement for a value to be run against to determine if it's an outlier
    upper_q_df[f'{treatment}'] = upper_bound.tolist()
    lower_q_df[f'{treatment}'] = lower_bound.tolist()

# For testing purposes, show which values are outliers in both directions for all 4 medicines. 
outlier1 = select_treatments_df["Capomulin"] < lower_q_df["Capomulin"]
outlier2 = select_treatments_df["Ramicane"] < lower_q_df["Ramicane"]
outlier3 = select_treatments_df["Infubinol"] < lower_q_df["Infubinol"]
outlier4 = select_treatments_df["Ceftamin"] < lower_q_df["Ceftamin"]
outlier5 = select_treatments_df["Capomulin"] > upper_q_df["Capomulin"]
outlier6 = select_treatments_df["Ramicane"] > upper_q_df["Ramicane"]
outlier7 = select_treatments_df["Infubinol"] > upper_q_df["Infubinol"]
outlier8 = select_treatments_df["Ceftamin"] > upper_q_df["Ceftamin"]
select_treatments_df
```

```python
# Generate a box plot of the final tumor volume of each mouse across four regimens of interest
# Creates a boxplot to map the tumor sizes. Marks outliers in red. 
fig1, ax1 = plt.subplots()
ax1.boxplot(select_treatments_df, flierprops = {'markerfacecolor':'r'})
# Assigns x ticks, title, and labels 
plt.xticks([1,2,3,4],treatment_list)
plt.title("Tumor Volume By Drug Regimen")
plt.xlabel("Drug Regimens")
plt.ylabel("Tumor Volume (mm3)")
plt.show()
```

```python
# Generate a scatter plot of average tumor volume vs. mouse weight for the Capomulin regimen
##  Need to create a DF that only has the final tumor size values when the treatment is Capomulin.
## final_tumor_size_df_capomulin = final_tumor_size_df.loc[final_tumor_size_df["Drug Regimen"] == "Capomulin"]

# Need to make a DF stores the average values for weight and tumor size when the mouse is treated with Capomulin
# Start with an empty DF to store those average values into
avg_tumor_size_capomulin = pd.DataFrame()

# This groupby allows us to extract the average for both weight and tumor size by mouse ID by pulling from the DF
# with mice only created via Capomulin
grouped_capomulin_df = treated_with_capomulin.groupby("Mouse ID")

# Use the same math functions from above to generate columns of data for the empty DF for average
# weight and tumor size
avg_tumor_size_capomulin["Average Weight (g)"] = grouped_capomulin_df["Weight (g)"].mean()
avg_tumor_size_capomulin["Average Tumor Volume (mm3)"] = grouped_capomulin_df["Tumor Volume (mm3)"].mean()


# Take the new DF and plot the average weights on the X axis and tumor size on y axis.
# Weights appear to be consistent for each mouse, but taking the average was the easiest way to get the values
plt.scatter(avg_tumor_size_capomulin["Average Weight (g)"], avg_tumor_size_capomulin["Average Tumor Volume (mm3)"])
plt.xlabel("Weight (g)")
plt.ylabel("Tumor Volume (mm3)")
plt.title("Average Weight (g) vs Tumor Volume (mm3) for Mice treated with Capomulin")
plt.show()

# treated_with_capomulin.head()
## final_tumor_size_df
## final_tumor_size_df_capomulin
# grouped_capomulin_df.head()
avg_tumor_size_capomulin.head()
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and linear regression model 
# for mouse weight and average tumor volume for the Capomulin regimen

# Set my X and Y values to be the same as the scatter plot above. 
x_values = avg_tumor_size_capomulin["Average Weight (g)"]
y_values = avg_tumor_size_capomulin["Average Tumor Volume (mm3)"]

slope, intercept, rvalue, pvalue, stderr = st.linregress(x_values, y_values)

regression_values = (x_values * slope) + intercept
correlation = st.pearsonr(x_values,y_values)[0]
plt.scatter(x_values,y_values, label = "Data Points")
plt.plot(x_values,regression_values,color="r", label = "Fitted Line")
plt.xlabel("Weight (g)")
plt.ylabel("Tumor Volume (mm3)")
plt.title("Average Weight (g) vs Tumor Volume (mm3) for Mice treated with Capomulin")
plt.legend(loc="best")
plt.show()
print(correlation)
```
